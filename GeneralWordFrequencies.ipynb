```python
# pandas
import pandas as pd

# spacy
import spacy
```

```python
nlp = spacy.load('en_core_web_sm', disable=['ner,parser'])
nlp.remove_pipe('ner')
nlp.remove_pipe('parser')
```

```python
'''
Read in .tsv of tagged sample data as a Pandas data frame
Add appropriate header to the columns as well.
'''
def read_as_df(filename):
    df = pd.read_csv(filename, sep="\t", header = None)
    df.columns = ["Text_Loc", "Sample", "Rating", "Specificity","Adj", "Adv", "Noun", "Verb", "Adp"]
    return df
```

```python
def read_word_list(filename):
    df = pd.read_csv(filename)
    df.columns = ['Words']
    return df
```

```python
def time_freq(df, word_list, word_category):
    counts = []
    for sample in df['Sample']:
        count = 0
        doc = nlp(sample)
        for word in doc:
            if word.text in word_list:
                count += 1
        counts.append(count)
    
    df[word_category] = counts
    return df
```

```python
time_words = read_word_list("data/time_words_hiq.csv")
```

```python
sample_df = read_as_df("data/samples_data.tsv")
```

```python
time_freq(sample_df, list(time_words['Words']), 'Time')
```
