```python
# read files
from os import listdir
from os.path import isfile, join

# sorting
import operator
```

```python
def aggregate_attention_data(directory_name):
    
    files = [f for f in listdir(directory_name) if isfile(join(directory_name, f))]
    term_dict = {}
    
    for file in files:
        if "dev_7.txt" in file:
            with open(directory_name + file, encoding="utf-8") as f:
                for line in f:
                    line = line[1:-2]
                    values = line.split(", ")
                    word = values[0][1:-1]
                    weight = float(values[1])

                    if word not in term_dict:
                        term_dict[word] = weight
                    else:
                        term_dict[word] += weight
        else:
            with open(directory_name + file, encoding="utf-8") as f:
                for line in f:
                    line = line[2:-3]
                    values = line.split(", ")
                    word = values[0][:-1]
                    weight = float(values[1])
                    if word != "[CLS]":
                        word = word[2:]
                    if word not in term_dict:
                        term_dict[word] = weight
                    else:
                        term_dict[word] += weight

    return term_dict
```

```python
info = aggregate_attention_data("data/attended_to_samples/")
```

```python
info_sorted = dict(sorted(info.items(), key=operator.itemgetter(1), reverse = True))
```

```python
info_sorted_low_to_high = dict(sorted(info.items(), key=operator.itemgetter(1), reverse = False))
```

```python
count = 0
print("TOP 100 TOKENS")
for k, v in info_sorted.items():
    print(f'{k}: {v}')
    count+=1
    if count == 100:
        break
```

```python
count = 0
print("BOTTOM 100 TOKENS")
for k, v in info_sorted_low_to_high.items():
    print(f'{k}: {v}')
    count+=1
    if count == 100:
        break
```
